```python
plt.figure(figsize=(15,8))
plt.pie(data['Sub-Category'].value_counts(),labels=data['Sub-Category'].value_counts().index,shadow=True,autopct="%0.1f%%")
plt.show()
```

```python
data['State'].value_counts()
```

```python


plt.figure(figsize=(15,12))
sns.countplot(x='State',data=data,palette='rocket_r',order=data['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

highest number of buyers are from callifornia and Newyork

```python
data.hist(figsize=(10,10),bins=50)
plt.show()
```

maximum customers tends to buy quantity of 2 & 3,

```python
plt.figure(figsize=(10,8))
data['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

```python
data.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['pink','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

profit and sales are maximum in consumer segment
minimum in home segment

```python

data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

profit and sales are maximum in west region and minimum in south region

```python

ps = data.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

high profit is for california, new york and loss is for texas, pennsylvania, Ohi

```python
data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','purple'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# As a business manager, try to find out the weak areas where you can work to make more profit?

Technology and Office Supplies have high profit.

Furniture have less profit

```python
ps = data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightblue'],figsize=(17,16))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

Phones sub-category have high sales.
chairs have high sales but less profit compared to phones
Tables and Bookmarks sub-categories facing huge loss
